````python
import random
import math



class MyLinearRegression:


    def __init__(self,filePath,fileName,ratio):
        self.coef = None
        self.intercept = None
        self.dataList = []
        self.trainDataSet = []
        self.testDataSet = []
        self.filePath = filePath
        self.fileName = fileName
        self.ratio = ratio
        self.RMSE = None
        self.test = None
        self.yPredictions = None
        self.yActual = None
        


    def loadData(self):
        file_path = self.filePath+self.fileName

        file = open(file_path, "r")

        listOfLists = []

        for line in file.read().splitlines():
            list_of_floats = [float(data_from_line) for data_from_line in line.split(",")]
            listOfLists.append(list_of_floats)
        file.close()
        self.dataList = listOfLists

    def sumList(self,aList):
        sum = 0
        for x in aList:
            sum = sum + x
        return sum

    def meanList(self,aList):
        mean = 0
        sum = self.sumList(aList)
        mean = sum/len(aList)
        return mean

    def minMaxList(aList):
        max = 0
        for x in aList:
            if x > max:
                max = x
        min = max
        for x in aList:
            if x < min:
                min = x
        return [min,max]




    def dataSplit(self):
        dataList = self.dataList
        trainSet = []
        trainSetSize = len(dataList) * self.ratio
        trainSetSize = int(trainSetSize)
        usedIndex = []
        indexList = range(0, trainSetSize)
        while len(trainSet) < trainSetSize:
            randomIndex = random.choice(indexList)
            if randomIndex not in usedIndex:
                trainSet.append(dataList[randomIndex])
                usedIndex.append(randomIndex)
        
        testSet = []
        testSetSize = len(dataList) - trainSetSize
        testIndexList = range(0, len(dataList))
        while len(testSet) < testSetSize:
            testRandomIndex = random.choice(testIndexList)
            if testRandomIndex not in usedIndex:
                testSet.append(dataList[testRandomIndex])
                usedIndex.append(testRandomIndex)
        self.trainDataSet = trainSet
        self.testDataSet = testSet


    def splitXYList(self,xYList):
        xList = []
        yList = []
        for index in range(0, len(xYList)):
            xList.append(xYList[index][0])
            yList.append(xYList[index][1])
        return xList,yList
    
    def covarianceList(self,listOfLists):
        xList = self.splitXYList(listOfLists)[0]
        yList = self.splitXYList(listOfLists)[1]
        meanOfX = self.meanList(xList)
        meanOfY = self.meanList(yList)
        subX = [i - meanOfX for i in xList]
        subY = [i - meanOfY for i in yList]

        covariance = self.meanList([subX[i]*subY[i] for i in range(len(subX))])
        return covariance
    
    def varianceList(self,listOfLists):
        xList = self.splitXYList(listOfLists)[0]

        mean = self.meanList(xList)

        tempList = [(x-mean) ** 2 for x in xList]

        variance = self.meanList(tempList)
        return variance


    def fit(self):
        trainingSet = self.trainDataSet
        xList = self.splitXYList(trainingSet)[0]
        yList = self.splitXYList(trainingSet)[1]
        xMean = self.meanList(xList)
        yMean = self.meanList(yList)
        covariance = self.covarianceList(trainingSet)
        variance = self.varianceList(trainingSet)
        bOne = covariance/variance
        bZero = yMean-bOne*xMean
        self.intercept = bZero
        self.coef = bOne
        return (bZero,bOne)


    def eval(self):
        predict = self.yPredictions
        actual = self.yActual
        sumOfError = 0
        for i in range(len(actual)):
            predictError = predict[i] - actual[i]
            sumOfError += (predictError**2)
        meanError = sumOfError/len(actual)
        self.RMSE = math.sqrt(meanError)



    def predictYTestSet(self):
        bZero = self.intercept
        bOne = self.coef
        testX = self.splitXYList(self.testDataSet)[0]
        testY = self.splitXYList(self.testDataSet)[1]
        yPredictions = []
        yActual = []
        for index in range(0,len(testX)):
            yPredictions.append(bZero+bOne*testX[index])
            yActual.append(testY[index])
        self.yPredictions = yPredictions
        self.yActual = yActual



    def output(self):
        self.loadData()
        self.dataSplit()
        self.fit()
        self.predictYTestSet()
        self.eval() 
        print("The coefficients b0 is %s; b1 is %s;RMSE is %s" % ("{:.2f}".format(self.intercept),"{:.2f}".format(self.coef),"{:.2f}".format(self.RMSE)))

regClassObject = MyLinearRegression("C:\Users\itstudent\Desktop\GTA\IT 170 Scripting Lanuage and Automation\prog 4,"salary.csv",.75)


regClassObject.output()
````
